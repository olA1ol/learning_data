```python
import pandas as pd
```

```python
##파일 로드

df = pd.read_csv("../data/csv/Sales Records.csv")
```

```python
df.head(1)
```

```python
## 특정 칼럼을 제거
df.drop(
    ['Order ID', 'Total Revenue', 'Total Cost',  'Total Cost', 'Total Profit', 'Sales Channel'], axis=1, inplace=True)
```

```python
df
```

```python
## 컬럼의 이름을 변경
df.columns = ['대륙', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가' ]
```

```python
df.head(1)
```

```python
##대륙, 국가를 기준으로 오름차순 정렬

df.sort_values(['대륙', '국가'], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
##새로운 파생 변수 생성
df['총단가'] = df['판매단위'] * df['단가']
df['총원가'] = df['판매단위'] * df['원가']
df['총이윤'] = df['총단가'] - df['총원가']
```

```python
df.head(3)
```

```python
## 대륙별 빈도수를 출력
df['대륙'].value_counts()
```

```python
# 대륙별 총 이윤의 합계가 가장 높은 대륙은 어디인가?

# 데이터프레임의 그룹화
# groupby({그룹화할 컬럼의 이름}).연산식
group_data = df.groupby('대륙')
```

```python
group_data
```

```python
group_data.sum()

#문자들은 다 연결이 된 채로 나타남.
#숫자는 합해진 값이 나타남.
```

```python
group_data.mean() #문자열 때문에 에러가 난다. -> 그룹화할 때 잘라서 넣어야...
```

```python
pd.pivot_table(
    df,
    index= "대륙",   
    columns= '상품종류',
    values= "총이윤",
    aggfunc= ["sum", "mean"]
)
```

```python
pd.pivot_table(
    df,
    index= "대륙",
    
    columns= '상품종류',
    values= "총이윤",
    aggfunc= "sum",
    margins= True ##상품종류별 총합계도 추가하는 방법.
)
```

```python
group_data4 = df[['대륙', '상품종류', '총이윤']].groupby(['대륙','상품종류']).sum()
group_data4
```

```python
#여기서 인덱스를 칼럼으로 올리기

group_data4.unstack() #최우측에 있는 인덱스 1개가 컬럼으로 올라감
```

```python
#여기서 인덱스를 칼럼으로 한 번 더 올리기

group_data4.unstack().unstack()
```
