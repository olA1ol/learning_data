## 연습문제
1. csv 폴더에 있는 Sales Records.csv 파일을 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 제거
3. 컬럼의 이름을 변경 -> ['대륙', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 기존의 인덱스는 제거하고 인덱스를 재지정
6. 새로운 파생변수 생성
    - 총단가 -> 판매단위 * 단가
    - 총원가 -> 판매단위 * 원가
    - 총이윤 -> 총단가 - 총원가
    

```python
import pandas as pd
```

```python
def sales_report(_df, _col):
    result = _df.copy()

    if "'Order ID: 0', 'Total Revenue: 0', 'Total Cost: 0', 'Total Profit: 0', 'Sales Channel: 0'" in result.columns:
        result.drop("'Order ID: 0', 'Total Revenue: 0', 'Total Cost: 0', 'Total Profit: 0', 'Sales Channel: 0'", axis=1)

    result.columns = ['대륙', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']

    result.sort_values('대륙', '국가')

    result.reset_index(drop=True)

    result['총단가'] = (result['판매단위'] * result['단가']).fillna()
    result['총원가'] = (result['판매단위'] * result['원가']).fillna()
    result['총이윤'] = (result['총단가'] - result['총원가']).fillna()

    return result
```
