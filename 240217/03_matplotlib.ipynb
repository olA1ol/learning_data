```python
# 라이브러리 로드
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt 라고 써도 됨
```

```python
## 라인 그래프 표시
## plot({1차원 데이터})
## 1차원 데이터를 인자 값에 하나 대입 : y축 데이터 입력, x축 데이터는 0부터 1씩 증가하는 데이터
## 1차원 데이터를 인자 값에 2개 대입 : 첫 번째 인자 x축 데이터, 두 번째 인자 y축 데이터

plt.plot([5,3,7,1])
plt.show
```

```python
plt.plot([1,2,3,4],[5,3,7,1])
plt.plot([1,3,5,7],[1,7,3,5])
```

```python
## 라인그래프의 선의 색을 변경
plt.plot([3,5,2,1], '#FF0066') #hex 값도 가능, rgb 값도 가능
plt.show()
```

```python
# 라인그래프 선의 종류
plt.plot([1,2,3,4], '--r')
plt.show()
```

```python
# x, y축의 라벨 설정
# 그래프의 제목 추가
plt.plot([1,2,3,4], [6,3,2,8])
plt.xlabel('A')
plt.ylabel('B')
plt.title('Graph')
plt.show()
```

```python
# 막대 그래프
# bar({x축 데이터}, {y축 데이터}) 둘 다 넣어줘야 함
x = [1,2,3,4]
y = [10, 40, 30, 70]

plt.bar(x, y)
plt.show()
```

```python
plt.barh(x, y) #수평 막대 그래프 barh
plt.show()
```

```python
plt.bar(x, y, color=['red', 'blue', 'yellow', 'green'], width=1) # 개수가 다른 경우, 로테이션하며 돌아간다
plt.show()
```

```python
plt.bar(x, y, color=['yellow', 'green'], width=0.3) # 개수가 다른 경우, 로테이션하며 돌아간다 #너비 1은 100%, 아주 얇게도 조정 가능
plt.show()
```

```python
plt.bar(['a', 'b', 'c', 'd'], y)
```

```python
plt.bar(['가', '나', '다', '라'], y) #문자열도 가능한데 한글은 에러 메시지가 뜨면서 깨진다.
```

```python
#맥플롯에서 지원하는 서체가 한글을 지원하지 않아서임.
#한글 폰트를 지정해주면 됨.
```
