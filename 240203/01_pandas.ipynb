### 데이터프레임 필터
조건을 기준으로 데이터를 볼 것이냐
위치를 기준으로 데이터를 볼 것이냐
두 가지로 설정할 수가 있음.

1. loc[{인덱스의 조건}, {컬럼의 조건}] (location)
    - 인덱스의 값을 기준으로 데이터프레임 필터
    - 컬럼의 값을 기준으로 데이터프레임 필터
    - 인덱스와 컬럼의 조건식을 이용하여 데이터프레임 필터
2. iloc[{인덱스의 위치}, {컬럼의 위치}] (integer location)
    - 인덱스나 컬럼의 위치를 기준으로 데이터프레임 필터

```python
# 데이터프레임에서 특정한 데이터를 제거
# drop( {조건} , axis = {0(rows) | 1(columns)}, inplace = {bool} )
# 데이터프레임명.drop()
# axis : 기준 -> 행{0}을 지울 것이냐?  열{1}을 지울 것이냐?
# inplace : 참 / 거짓 형태의 데이터를 입력
# 참 : 기준이 되는 데이터프레임을 변경
# 거짓 : 기준이 되는 데이터프레임을 변경하지 않고 결과를 출력 (기본값)

# loc와 iloc의 차이를 보여주기 위한.. 부가 조건들

# df에서 인덱스 0인 데이터를 제거
# df에서 Unnamed: 0 컬럼을 제거 - "콜론하고 한 칸 띄고 0"
df.drop('Unnamed: 0', axis=1)
# 해당 컬럼의 이름이 없을 때 자동으로 Unnamed가 됨..
```

```python
# inplace = True를 이해하기 위한 풀어쓰기 과정

class Test:
    def __init__(self, data, index = None, columns = None):
        self.data = data
        if index:
            self.index = index
        else:
            self.index = list(range(len(data)))
        if columns:
            self.columns = columns
        else:
            self.columns = list(range(len(data[0])))
    def __repr__(self):
        return str(self.data)   #repr이라는 함수를 쓰려면 문자열이어야 해서 str을 적어준 것.
    
    def add(self, value, inplace = False):
        if inplace:
            self.data = self.data + [value]
        else:
            return self.data + [value]
```
